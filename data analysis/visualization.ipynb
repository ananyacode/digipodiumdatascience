```python
x=np.arange(-5,5,0.2)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15,5))
axs[0].plot(x,y1,'r--')
axs[1].plot(x,y2,'b--')
axs[2].plot(x,y3,'g--')
axs[0].set_title('sin(x)')
axs[1].set_title('cos(x)')
axs[2].set_title('tan(x)')
plt.show()
```

implicit vs explicit graphs

```python
# implict figure and axes creation
x = np.linspace(-10,10,25)
plt.plot(x,x,'-',label = 'linear')
plt.plot(x,x**2,'--',label = 'quadratic')
plt.plot(x,x**3,'o:',label = 'cubic')
plt.xlabel('my x label')
plt.ylabel('my y label')
plt.title('Simple graph with implicit graph setting')
plt.legend()
plt.show()
plt.savefig('Plot1.png',bbox_inches='tight',dpi=300)
```

```python
x = np.linspace(-10,10,25)
fig,ax=plt.subplots()
ax.plot(x,x,'g-',label = 'linear')
ax.plot(x,x**2,'r--',label = 'quadratic')
ax.plot(x,x**3,'o:',label = 'cubic')
ax.set_xlabel('x label')
ax.set_xlabel('y label')
ax.set_title('simple plot')
ax.legend()
plt.show()
plt.savefig('Plot2.png',bbox_inches='tight',dpi=300)
```

```python
data = {
    "a": np.arange(50),
    "b": np.random.randint(0,5,50),
    "c": np.random.randn(50),
    "d": np.arange(50)+10*np.random.randn(50)
}
data['d'] = np.abs(data['d'])*10
fig, ax= plt.subplots(figsize=(10,3))
ax.scatter('a','b',c='c',s='d',data='data')
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
ax.set_title('scatter plot')
plt.show()
```

```python
data = {
    'names' : ['bob' , 'jessica', 'mary', 'john', 'mel'],
    'english' : [76,77,78,73,70],
    'maths' : [45,76,87,78,56],
    'hindi' : [71,81,23,54,52],
}
```

```python
def my_custom_plotter(ax,x_value,y_value,param_dict):
    out = ax.bar(x_value,y_value,**param_dict)
    ax.set_title('visualizing marks')
    ax.set_xlabel("marks")
    ax.set_ylabel("count")
    return out
```

```python
fig,axs= plt.subplots(1,3,figsize= (15,3))
my_custom_plotter(axs[0],'names','english',{'data':data,'color':'red','alpha':.3})
my_custom_plotter(axs[1],'names','english',{'data':data,'color':'green','alpha':.3})
my_custom_plotter(axs[2],'names','maths',{'data':data})
```

Plotting multiple lines

```python
years = list(range(2000,2012))
apples = [0.895,0.91,0.919,0.926,0.929,0.931,0.934,0.936,0.937,0.9375,0.9372,0.939]
oranges = [0.962,0.941,0.930,0.932,0.918,0.908,0.904,0.901,0.898,0.9,0.896,0.891]
```

```python
plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel('Years')
plt.ylabel('Yield (tons per hactare)');
```

Chart title and legend

```python
plt.plot(years,apples,label='Apples')
plt.plot(years,oranges, label='Oranges')

plt.xlabel('Years')
plt.ylabel('Yield (tons per hactare)');

plt.title('Crop yields in kanto')
plt.legend();
```

Line Markers

```python
plt.plot(years,apples,label='Apples',marker = 'o')
plt.plot(years,oranges, label='Oranges',marker ='x')

plt.xlabel('Years')
plt.ylabel('Yield (tons per hactare)');

plt.title('Crop yields in kanto')
plt.legend();
```

Studing lines and markers

- color or c - to set the color of line
- linestyle or ls : choose between a solid or dashed line
- linewidth or lw : set width of the line
- markersize or ms : set the size of marker
- markeredgecolor or mec : set the edge colors for markers
- markerfacecolor or mfc : set the fil color for markers
- alpha : opacity of the plot
- FMTcommon linestyles you can use
    - '-' (solid line, default)
    - '--' (dashed line)
    - ':' (dotted line)
    - '-.' (dash-dot line)

```python
plt.plot(years,apples,'s-b',label='Apples',lw = 2,ms=8,mew=2,mec='navy')
plt.plot(years,oranges,'o--r', label='Oranges',lw = 3,ms=10,alpha =0.5)

plt.xlabel('Years')
plt.ylabel('Yield (tons per hactare)');

plt.title('Crop yields in kanto')
plt.legend();
```

```python
plt.plot(years,apples,'sb',label='Apples',lw = 2,ms=8,mew=2,mec='navy')
plt.plot(years,oranges,'or', label='Oranges',lw = 3,ms=10,alpha =0.5)

plt.xlabel('Years')
plt.ylabel('Yield (tons per hactare)');

plt.title('Crop yields in kanto')
plt.legend();
```
