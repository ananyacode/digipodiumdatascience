```python
import matplotlib.pyplot as plt # basic graph
import numpy as np 
import pandas as pd
import seaborn as sns # advanced graph
import plotly.express as px # interactive graphs
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
print(titanic['survived'].unique())
print(titanic['pclass'].unique())
print(titanic['age'].unique())
```

# categorial graphs
- columns that have text are generally categorial
- the first step is to count the number of unique values in column
- the plot the count of each category
- following charts are a good way to visualize data
  - bar chart
  - pie chart
  - donut chart
  - line chart
  - area chart

```python
survival_count = titanic['survived'].value_counts()
survival_count.plot(kind= 'pie',autopct='%.1f%%',title = 'Survival Rate' ); # autopct for percentage 
```

```python
survival_count.plot(
    kind = 'pie',
    autopct = '%.1f%%',
    title = 'Survival Rate',
    labels = ['Survived','Not Survived'],
    colors = ['#66b3ff','#ff9999'],
    labeldistance = 1.1,
    wedgeprops = {
        'width': 0.5,
    }
);
```

```python
sex_category = titanic['sex'].value_counts()
sex_category.plot(
    kind = 'pie',
    autopct = '%.1f%%',
    title = 'Sex',
    labels = ['Male','Female'],
    colors = ['#66b3ff','#ff9999'],
    labeldistance = 1.1,
    wedgeprops = {
        'width': 0.5,
    }
);
```

```python
pclass_category = titanic['pclass'].value_counts()
pclass_category.plot(
    kind = 'pie',
    autopct = '%.1f%%',
    title = 'pclass',
    labels = ['1','2','3'],
    colors = ['#66b3ff','#ff9999','#99ff99'],
    labeldistance = 1.1,
    wedgeprops = {
        'width': 0.5,
    }
);
```
