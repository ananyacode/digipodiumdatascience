```python
iris = sns.load_dataset('iris')
iris
```

```python
sns.set_theme()
sns.relplot(
    data=iris,
    x="sepal_length", y="sepal_width",col='species',
    hue="species"
);
```

```python
#the col="species" parameter creates separate columns for each species of iris flowers
#hue="species" parameter color-codes the data points within each species

sns.relplot(
    data=iris, kind="line",
    x="petal_length", y="petal_width", col="species",  
    hue="species", style="species",
    facet_kws=dict(sharex=False),
)
```

```python
sns.relplot(
    data=iris, kind="line",
    x="sepal_length", y="petal_length", col="species",
    hue="species", style="species",
)
```

The regplot function combines a scatter plot with a linear regression model to estimate and display the linear relationship between the variables

The lmplot allows you to create multiple scatter plots with regression lines, each representing a different level of a categorical variable


FacetGrid is a powerful feature in Seaborn that allows you to create a grid of subplots (facets) based on one or more categorical variables

lmplot = regplot + facetgrid

```python
sns.lmplot(data=iris, x="sepal_length", y="sepal_width", col="species", hue="species")
```

-lmplot
Combine regplot() and FacetGrid to plot multiple linear relationships in a dataset.

jointplot
Combine regplot() and JointGrid (when used with kind="reg").

pairplot
Combine regplot() and PairGrid (when used with kind="reg").

residplot
Plot the residuals of a linear regression model.

```python
sns.pairplot(data=iris, hue="species")
```

```python
g = sns.PairGrid(iris, hue="species", corner=True)
g.map_lower(sns.kdeplot, hue=None, levels=5, color=".2")
g.map_lower(sns.scatterplot, marker="+")
g.map_diag(sns.histplot, element="step", linewidth=0, kde=True)
g.add_legend(frameon=True)
g.legend.set_bbox_to_anchor((.61, .6))
```

```python
sns.set_theme(style="ticks", font_scale=1.25)
g = sns.relplot(
    data=iris,
    x="sepal_length", y="sepal_width", hue="species",
    palette="crest", marker="x", s=100,
)
g.set_axis_labels("sepal_length", "sepal_width", labelpad=10)
g.legend.set_title("iris")
g.figure.set_size_inches(6.5, 4.5)
g.ax.margins(.15)
g.despine(trim=True)
```
